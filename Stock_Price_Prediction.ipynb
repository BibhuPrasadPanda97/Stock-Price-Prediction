# Stock Price Prediction

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import date
import nsepy

# import warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# import data of HDFCBANK stock
hdfc_data = nsepy.get_history('HDFCBANK', start=date(2015,1,1), end=date(2022,5,16))
hdfc_data.head()
```

```python
hdfc_data.tail(10)
```

```python
# plot close price of the stock
plt.figure(figsize=[15,6])
plt.plot(hdfc_data['Close'], label='HDFCBANK stocks price')
plt.legend()
plt.title("HDFCBANK", fontsize=15)
plt.show()
```

## Moving Average for 100 day window

```python
# moving average
y_pred_ma = hdfc_data['Close'].rolling(30).mean()

# plot
plt.figure(figsize=[15,6])
plt.plot(hdfc_data['Close'], label='HDFCBANK stocks price')
plt.plot(y_pred_ma, label='Moving average of HDFCBANK', color='r')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Price')
plt.title("HDFCBANK", fontsize=15)
plt.show()
```
